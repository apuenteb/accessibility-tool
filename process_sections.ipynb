Notebook to generate census sections with colors

```python
import geopandas as gpd

sections=gpd.read_file('assets/geojsons/sections_colores.geojson')
```

```python
print(sections.columns)
```

```python
# rename color_1_walk to color
sections.rename(columns={'color_1_walk':'color'}, inplace=True)
print(sections.columns)
```

```python
import matplotlib.pyplot as plt

# Create figure with black background
fig, ax = plt.subplots(figsize=(10, 10), facecolor='black')
ax.set_facecolor('black')

# Plot the geometries with colors from the 'color' column
sections.plot(ax=ax, color=sections["color"], edgecolor='none')

# Remove axis
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Save the image
plt.savefig("geojson_plot.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```
